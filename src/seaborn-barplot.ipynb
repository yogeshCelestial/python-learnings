```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
flights = sb.load_dataset('flights')
flights.describe()
```

## Bar Plot 

```python
plt.figure(figsize=(16, 6))
sb.barplot(data=flights, x='year', y='passengers', hue='month')
```

### Number of passangers in flights by year growth

```python
sb.barplot(data=flights, x='month', y='passengers', estimator='max', errcolor='white')
```

### As shown in bar graph, July and August is the peak of the season
