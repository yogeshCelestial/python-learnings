```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
flights = sb.load_dataset('flights')
flights.head()
```

```python
flights_pivot = flights.pivot(index='year', columns='month', values='passengers')
print(flights_pivot)
plt.figure(figsize=(20,6))
sb.heatmap(flights_pivot, annot=True, fmt='d')
```

## Line color and line width

```python
plt.figure(figsize=(20,6))
sb.heatmap(flights_pivot, annot=True, fmt='d', linecolor='black', linewidths='2', cbar_kws={ 'orientation': 'horizontal' })
```

## Cmap

```python
plt.figure(figsize=(20,6))
sb.heatmap(flights_pivot, annot=True, fmt='d', linecolor='black', linewidths='2', cmap='Blues')
```
