```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

# Scatter Plot Graph

```python
sales_info = pd.read_csv('./assets/car_sales.csv')
sales_info.head()
```

```python
sb.scatterplot(x='year', y='sales', data=sales_info)
```

```python
titanic_data = sb.load_dataset('titanic')
titanic_data.head()
```

```python
sb.scatterplot(x='age', y='fare', data=titanic_data, hue='alive')
```

```python
plt.figure(figsize=(16,6))
sb.scatterplot(x='age', y='fare', data=titanic_data, hue='alive', style='class')
```

### the above two graphs shows that travellers who are from second and first class are survived more than third call travellers

```python
plt.figure(figsize=(16,6))
sb.scatterplot(x='age', y='fare', data=titanic_data, hue='alive', style='class')
sb.lineplot(x='age', y='fare', data=titanic_data.head(100), hue='alive', style='class')

plt.title('Scatter and Line plot')
```
