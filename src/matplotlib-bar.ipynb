```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
plt.style.use('dark_background')
```

```python
# Data
categories = ['Apples', 'Bananas', 'Oranges', 'Grapes']
values = [45, 30, 55, 40]

# Create plot
plt.figure(figsize=(8, 5))
bars = plt.bar(categories, values,
               color='skyblue',
               edgecolor='black',
               linewidth=1.5,
               width=0.6,
               alpha=0.8)

# Customize
plt.title('Fruit Sales', fontsize=14)
plt.xlabel('Fruits', fontsize=12)
plt.ylabel('Quantity Sold', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}',
             ha='center', va='bottom')

plt.show()
```

```python
# Data
products = ['Laptops', 'Phones', 'Tablets', 'Watches']
sales = [120, 85, 60, 45]
errors = [10, 5, 8, 3]

# Sort data for better visualization
sorted_idx = np.argsort(sales)
products = [products[i] for i in sorted_idx]
sales = [sales[i] for i in sorted_idx]
errors = [errors[i] for i in sorted_idx]

# Create plot
plt.figure(figsize=(9, 5))
bars = plt.barh(products, sales,
                xerr=errors,
                color='lightgreen',
                edgecolor='darkgreen',
                height=0.6,
                capsize=5,
                alpha=0.8)

# Customize
plt.title('Product Sales with Error Bars', fontsize=14)
plt.xlabel('Units Sold', fontsize=12)
plt.ylabel('Products', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.xlim(0, 140)
plt.tight_layout()

plt.show()
```

```python
# Data
categories = ['Q1', 'Q2', 'Q3', 'Q4']
sales_2022 = [80, 95, 70, 85]
sales_2023 = [90, 85, 75, 95]

# Set positions and width
x = np.arange(len(categories))
print(x, 'xxxxxxxxxxxxxxxxxxxxxx')
width = 0.35

# Create plot
plt.figure(figsize=(9, 6))
bar1 = plt.bar(x - width/2, sales_2022, width,
               label='2022',
               color='salmon',
               edgecolor='black',
               hatch='//',
               alpha=0.8)

bar2 = plt.bar(x + width/2, sales_2023, width,
               label='2023',
               color='royalblue',
               edgecolor='black',
               hatch='\\\\',
               alpha=0.8)

# Customize
plt.title('Quarterly Sales Comparison', fontsize=14)
plt.xlabel('Quarters', fontsize=12)
plt.ylabel('Sales (Units)', fontsize=12)
plt.xticks(x, categories)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                 f'{height}',
                 ha='center', va='bottom')

add_labels(bar1)
add_labels(bar2)

plt.tight_layout()
plt.show()
```
