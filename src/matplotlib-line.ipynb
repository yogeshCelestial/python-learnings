```python
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
import pandas as pd
```

```python
# Generate data
x = np.linspace(0, 10, 20)
y = np.sin(x)

# Create plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, 
         marker='o',        # Add circular markers
         linestyle='--',    # Dashed line
         color='purple',
         markersize=8,
         markerfacecolor='yellow',
         markeredgecolor='black')

# Customize
plt.title("Sine Wave with Markers", fontsize=14)
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(['sin(x)'])
plt.show()
```

```python
# Generate data
x = np.arange(5)
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 9, 11]
errors = [0.5, 0.8, 0.3, 1.2, 0.6]

# Plot lines with error bars
plt.figure(figsize=(8, 4))
plt.errorbar(x, y1, yerr=errors, fmt='-s', label='Dataset 1', 
             color='green', ecolor='red', capsize=5)
plt.plot(x, y2, marker='^', linestyle=':', label='Dataset 2', color='orange')

# Customize
plt.title("Comparison with Error Bars", fontsize=14)
plt.xlabel("Index", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.xticks(x, ['A', 'B', 'C', 'D', 'E'])
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```
