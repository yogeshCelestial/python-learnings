```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
data1 = sb.load_dataset('titanic')
data1
```

# Line plotting for a Predefined dataSet

```python
sb.lineplot(x='age', y='survived', data=data1.head(100))
```

```python
marks = pd.read_csv('./assets/marks.csv')
marks.head()
```

```python
sb.lineplot(x='roll_number', y='english', data=marks)
```

```python
sales_info = pd.read_csv('./assets/car_sales.csv')
sales_info.head()
```

```python
sb.lineplot(x='year', y='sales', data=sales_info)
```

# Line Plot with user defined data frame from a csv file

```python
plt.figure(figsize=(20, 10))
sb.lineplot(x='country', y='sales', data=sales_info.head(25))
```

## Other Parameters in Line plot Fucntion

```python
plt.figure(figsize=(16, 8))
sb.lineplot(x='year', y='sales', data=sales_info.head(30), hue='country')
```

```python
plt.figure(figsize=(12, 6))
sb.lineplot(x='year', y='sales', data=sales_info.head(30), hue='country', style='country', legend='full', palette='Reds')
```
